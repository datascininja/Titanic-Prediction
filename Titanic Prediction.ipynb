### Imputation with Linear Regression

```python
import scipy.stats as stats
from sklearn import linear_model 

def regressionPred (traindata,testdata):

    indexed_test = testdata.set_index([np.arange(892,1310)])
    df = pd.concat([traindata,indexed_test])

    commonAdd = ['Miss','Mrs.','Master','Mr.','Rev']
    df['Name'][df['Name'].str.contains('Miss.')] = 'Miss'
    df['Name'][df['Name'].str.contains('Mrs.')] = 'Mrs.'
    df['Name'][df['Name'].str.contains('Master')] = 'Master'
    df['Name'][df['Name'].str.contains('Mr. ')] = 'Mr.'
    df['Name'][df['Name'].str.contains('Rev.')] = 'Rev'
    df['Name'][ ~df['Name'].isin(commonAdd) ] = 'Others'
    
    nameDummy = pd.get_dummies(df.Name)
    sexDummy = pd.get_dummies(df.Sex)
    df = df.join([sexDummy, nameDummy])
    
    df['Sex'][df['Sex'] == 'male'] = 1
    df['Sex'][df['Sex'] == 'female'] = 0 
    
    trainOn = df[~df.Age.isnull()]
    predictOn = df[df.Age.isnull()]
    y = trainOn.Age
    
    predictors = ['female','male','Pclass','SibSp','Parch','Master','Miss','Mr.','Mrs.','Others','Rev']
    regr = linear_model.LinearRegression()
    regr.fit(trainOn.ix[:,predictors], y)
    
    prediction = regr.predict( predictOn.ix[:,predictors])
    
    missingIndex = df[pd.isnull(df['Age'])].index 
    df.loc[missingIndex, 'Age'] = prediction
    
    return df

df = regressionPred(train,test); df.head() 
```

### Build a Logistic Regression Model

```python
predictors = ['Age','Pclass','Sex']
trainPred = df.ix[:891,predictors]
trainY = df.Survived[:891
                     ]

testID = (np.arange(892,len(df)+1))
testPred = df.ix[testID,predictors]

# Fit the Logistic Regression Model
log = LogisticRegression()
log.fit(trainPred,trainY)

# Predict on Test
predictions = log.predict(testPred)

# Concatenate
predDF = pd.concat([pd.DataFrame(testID), pd.DataFrame(predictions)], axis=1)
predDF.columns = ['PassengerID','Survived']

# Output Prediction as CSV
predDf = pd.DataFrame(predDF).to_csv('prediction3.csv')
```
